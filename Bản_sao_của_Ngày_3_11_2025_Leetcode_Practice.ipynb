# [1119] Remove vowels from string
* Một số điểm lưu ý:
1.  Tạo chuỗi mới 2 dấu nháy không khoảng trắng (+) hoặc tạo danh sách (.append())
2.  Khi kiểm tra chuyển letter thành kí tự để ko phân biệt hoa thường





```python

#đầu vào là 1 chuỗi
s=input()
#tạo chuỗi mới
new_list=""
# duyệt các kí tự thuộc chuỗi
for letter in s:
    # nếu không có kí tự trong set là {'a', 'e', 'i','o','u'} thì tiến hành nối thêm vào chuỗi mới
    if not letter.lower() in {'a', 'e', 'i','o','u'}:
            new_list+=letter
#in chuỗi mới
print (new_list)
```
